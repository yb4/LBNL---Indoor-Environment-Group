```python
#Sort columns of the DataTable
def sortDataTable(data):
    data['Time'] = pd.to_datetime(data['Time'], format='%m/%d/%Y %H%M')
    time = data["Time"].values
    forms = []
    pms = []
    intemps = []
    outtemps = []
    pm25s = []
    pm10s = []
    co2s = []
    inrhs = []
    outrhs = []
    ibus = []
    anms = []
    mtrs = []
    doors = []
    pm1s = [] #
    no2s = [] #
    plds = [] #
    ufps = [] #
    tvocs = [] #
    title_array = []
    #iterate each column (column name) in dataset
    for x in data[1:]:
        if x[0:3] == "FRM":
            forms += [x]
        elif x == 'PMin' or x == 'PMout':
            pms += [x]
        elif x == 'OUT_T' or x == 'OUT2_T' or x == 'CLR_OUT_T':
            outtemps += [x]
        elif x[0:2] == 'T_' or x[len(x)-2:] == '_T':
            intemps += [x]
        elif x == 'OUT_RH' or x == 'OUT2_RH' or x == 'CLR_OUT_RH':
            outrhs += [x]
        elif x[0:3] == 'RH_' or x[len(x)-3:] == '_RH':
            inrhs += [x]
        elif x[len(x)-5:] == '_PM10':
            pm10s += [x]
        elif x[len(x)-5:] == '_PM25':
            pm25s += [x]
        elif x[len(x)-4:] == '_PM1':
            pm1s += [x]
        elif x[0:4] == 'NO2_' or x[len(x)-3:] == 'NO2' :
            no2s += [x]
        elif x[0:4] == 'CO2_' or x[len(x)-3:] == 'CO2' :
            co2s += [x]
        elif x[0:4] == 'IBU_':
            ibus += [x]
        elif x[0:4] == 'ANM_':
            anms += [x]
        elif x[0:4] == 'MTR_':
            mtrs += [x]
        elif x[0:4] == 'STA_':
            doors += [x]
        elif x[0:4] == 'PLD_':
            plds += [x]
        elif x[0:4] == 'UFP_':
            ufps += [x]
        elif x[len(x)-4:] == 'tVOC':
            tvocs += [x]
        

    return (time, forms,pms,intemps,outtemps,pm1s,pm25s,pm10s,ufps,co2s,no2s,inrhs,outrhs,ibus,anms,mtrs,plds,doors,tvocs)
```
